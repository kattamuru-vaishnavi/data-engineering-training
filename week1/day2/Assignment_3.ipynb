```sql
%sql
--find the total salary for each department in the employee table
select department, sum(salary) as total_salary from Employee group by department;
```

```sql
%sql
--Count the number of employees in each department.
select department, count(*) as total_employees from Employee group by department;
```

```sql
%sql
--Calculate the average salary of employees in each department.
select department, avg(salary) as avg_salary from Employee group by department
```

```sql
%sql
--Find the maximum salary in each department.
select department, max(salary) as max_salary from Employee group by department;
```

```sql
%sql
--Find the minimum salary in each department.
select department, min(salary) as min_salary from Employee group by department;
```

```sql
%sql
--Find the total salary for departments where the total salary exceeds 100,000.
select department, sum(salary) as total_salary from Employee group by department having total_salary > 100000;
```

```sql
%sql
--Count the number of employees in departments with more than 2 employees.
select department, count(*) as total_employees from Employee group by department having total_employees > 2;
    
```

```sql

--16 total sales amount for each employee
select e.emp_name, sum(s.amount) as total_sales from employee e join sales s on e.emp_id = s.emp_id group by e.emp_name;
```

```sql


--17 number of sales made by each employee
select e.emp_name, count(s.sales_id) as sales_count from employee e join sales s on e.emp_id = s.emp_id group by e.emp_name;
```

```sql
%sql
--18 total sales amount grouped by product
select product, sum(amount) as total_sales from sales group by product;

```

```sql
--19 average sales amount grouped by product
select product, avg(amount) as avg_sales from sales group by product;
```

```sql
--20 employees who made more than 2 sales
select e.emp_name, count(s.sales_id) as sales_count from employee e join sales s on e.emp_id = s.emp_id group by e.emp_name having count(s.sales_id) > 2;
```

```sql
--21 total salary and total sales for each employee
select e.emp_name, sum(e.salary) as total_salary, sum(s.amount) as total_sales from employee e left join sales s on e.emp_id = s.emp_id group by e.emp_name;
```

```sql

--22 count unique products sold by each employee
select e.emp_name, count(distinct s.product) as unique_products from employee e join sales s on e.emp_id = s.emp_id group by e.emp_name;
```

```sql

--23 highest sales amount made by each employee
select e.emp_name, max(s.amount) as highest_sale from employee e join sales s on e.emp_id = s.emp_id group by e.emp_name;
```

```sql
--24 total sales grouped by product where total > 50000
select product, sum(amount) as total_sales from sales group by product having sum(amount) > 50000;
```

```sql

--25 departments with highest average sales amount
select e.department, avg(s.amount) as avg_sales from employee e join sales s on e.emp_id = s.emp_id group by e.department order by avg_sales desc limit 1;
```

```sql
%sql
--26 department with highest total sales amount
select e.department, sum(s.amount) as total_sales from employee e join sales s on e.emp_id = s.emp_id group by e.department order by total_sales desc limit 1;


```

```sql

--27 top 3 employees with highest total sales
select e.emp_name, sum(s.amount) as total_sales from employee e join sales s on e.emp_id = s.emp_id group by e.emp_name order by total_sales desc limit 3;
```
